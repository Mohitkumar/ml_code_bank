```python
import mxnet as mx
import numpy as np
```

```python
mnist = mx.test_utils.get_mnist()
batch_size = 100
```

```python
train_iter = mx.io.NDArrayIter(mnist['train_data'],mnist['train_label'],batch_size,shuffle=True)
val_iter = mx.io.NDArrayIter(mnist['test_data'], mnist['test_label'],shuffle=False, batch_size=batch_size)
```

```python
data = mx.sym.Variable('data')
data = mx.sym.flatten(data=data)
```

```python
fc1 = mx.sym.FullyConnected(data,num_hidden=128)
act1 = mx.sym.Activation(data=fc1, act_type='relu')
fc2 = mx.sym.FullyConnected(data=act1,num_hidden=64)
act2 = mx.sym.Activation(data=fc2, act_type='relu')
fc3 = mx.sym.FullyConnected(data=act2, num_hidden=10)
mlp = mx.sym.Softmax(data=fc3, name='softmax')
```

```python
model = mx.mod.Module(symbol=mlp, context=mx.cpu())
```

```python
import logging
logging.getLogger().setLevel(logging.DEBUG)
model.fit(train_data=train_iter,
          eval_data=val_iter, 
          optimizer='sgd',
          optimizer_params={'learning_rate':.01}, 
          batch_end_callback=mx.callback.Speedometer(batch_size, 100),num_epoch=50)
```

```python
test_iter = mx.io.NDArrayIter(mnist['test_data'], mnist['test_label'], batch_size)
acc = mx.metric.Accuracy()
model.score(test_iter,acc)
print acc
```
